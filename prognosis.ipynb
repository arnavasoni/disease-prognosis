### Methods

```python
# refer below, you're getting 5 symptoms inputs. while reading the symptoms, there should be a 1 for the same, and 0 for the remaining.
l2 = []
for x in range(0,len(symptoms)):
    l2.append(0)
```

```python
def NaiveBayes():
    from sklearn.naive_bayes import MultinomialNB
    gnb = MultinomialNB()
    gnb=gnb.fit(X_train.values,np.ravel(y_train))
    from sklearn.metrics import accuracy_score
    y_pred = gnb.predict(X_test.values)
    print(accuracy_score(y_test, y_pred))
    print(accuracy_score(y_test, y_pred, normalize=False))

    psymptoms = [Symptom1.get(),Symptom2.get(),Symptom3.get(),Symptom4.get(),Symptom5.get()]

    for k in range(0,len(symptoms)):
        for z in psymptoms:
            if(z==symptoms[k]):
                l2[k]=1
    # register the symptom in the checking list
    
    input_test = [l2]
    predict = gnb.predict(input_test)
    predicted=predict[0]

    h='no'
    for a in range(0,len(test_diseases)):
        if(test_diseases[predicted] == test_diseases[a]):
            h='yes'
            break

    if (h=='yes'):
        t3.delete("1.0", END)
        t3.insert(END, test_diseases[a])
    else:
        t3.delete("1.0", END)
        t3.insert(END, "No Disease")
```
